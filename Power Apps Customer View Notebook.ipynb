```python
# Load Silver and Scored tables
silver = spark.read.table("silver")
scored = spark.read.table("customer_churn_scored")

# Join and select relevant columns
df_console = (
    silver.join(scored, on="CustomerID", how="inner")
    .select(
        "CustomerID",
        "PhoneNumber",
        "State",
        "ContractType",
        "IntlPlan",
        "UnlimitedDataPlan",
        "MonthlyCharge",
        "TotalCharges",
        "CustomerServiceCalls",
        "ChurnRiskSegment",
        "ChurnProbability"
    )
)

display(df_console)
```

```python
# Save as a new table
df_console.write.mode("overwrite").saveAsTable("PowerApps_CustomerView")
```
