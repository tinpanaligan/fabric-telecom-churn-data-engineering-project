```python
# Load silver and scored tables
silver = spark.read.table("silver")
scored = spark.read.table("customer_churn_scored")

# Join on CustomerID
df = silver.join(scored, on="CustomerID", how="inner")

# Filter to high churn risk and 3+ service calls
df_support = df.filter(
    (df.ChurnRiskSegment == "High") &
    (df.CustomerServiceCalls >= 3)
)

# Select relevant support-focused fields
df_support = df_support.select(
    "CustomerID",
    "CustomerServiceCalls",
    "ContractType",
    "UnlimitedDataPlan",
    "IntlPlan",
    "ChurnRiskSegment",
    "ChurnProbability",
    "PhoneNumber",
    "State"
)

display(df_support)
```

```python
# Save as new table: Support_RiskWatch
df_support.write.mode("overwrite").saveAsTable("support_riskwatch")
```
