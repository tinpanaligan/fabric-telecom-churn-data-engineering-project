```python
from pyspark.sql.functions import count, avg, sum, when, col

# Load Silver table
df_silver = spark.read.table("silver")  # make sure this is the right table name

# Create Gold summary: churn & revenue by contract type
df_billing_gold = (
    df_silver.groupBy("ContractType")
    .agg(
        count("*").alias("TotalCustomers"),
        sum(when(col("ChurnLabel") == "Yes", 1).otherwise(0)).alias("ChurnedCustomers"),
        avg("MonthlyCharge").alias("AvgMonthlyCharge"),
        avg("TotalCharges").alias("AvgTotalCharges")
    )
)

# Add ChurnRate after aggregation
df_billing_gold = df_billing_gold.withColumn(
    "ChurnRate",
    col("ChurnedCustomers") / col("TotalCustomers")
)

display(df_billing_gold)

```

```python
# Save as Gold table
df_billing_gold.write.mode("overwrite").saveAsTable("Churn_Summary_ByContractType")
```
