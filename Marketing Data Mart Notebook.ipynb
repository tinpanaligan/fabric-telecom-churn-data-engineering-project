```python
# Load Silver and Scored tables
silver = spark.read.table("silver")
scored = spark.read.table("customer_churn_scored")

# Join them on CustomerID
df = silver.join(scored, on="CustomerID", how="inner")

# Filter to only Medium and High risk customers
df_marketing = df.filter(df.ChurnRiskSegment.isin("Medium", "High"))

# display(df_marketing)

# Select marketing-relevant fields
df_marketing = df_marketing.select(
    "CustomerID",
    "AgeGroup",
    "ContractType",
    "IntlPlan",
    "HighRevenueCustomer",
    "ChurnRiskSegment",
    "ChurnProbability",
    "PhoneNumber",
    "State"
)


display(df_marketing)
```

```python
# Save as a new table (Data Mart)
df_marketing.write.mode("overwrite").saveAsTable("marketing_segment")
```
